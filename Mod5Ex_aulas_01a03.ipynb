````python
import pandas as pd
import requests

# 1) seu código aqui
sinasc = pd.read_csv('SINASC_RO_2019.csv')
print(sinasc.shape)
sinasc.drop_duplicates().shape
# Não há duplicados
````

````python
# 2) seu código aqui
sinasc.isna().sum()
````

````python
# 3) seu código aqui
colunas = ['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO','GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5'] 
sinasc[colunas].isna().sum()
````

````python
# 4) seu código aqui
print(sinasc['APGAR5'].shape)
print(sinasc['APGAR5'].isna().sum())
````

````python
sinasc.dropna(subset=['APGAR5'], inplace=True)
````

````python
print(sinasc['APGAR5'].shape)
print(sinasc['APGAR5'].isna().sum())
````

````python
# 5) seu código aqui
sinasc[['ESTCIVMAE','CONSULTAS']] = sinasc[['ESTCIVMAE','CONSULTAS']].fillna(9)
sinasc[['ESTCIVMAE','CONSULTAS']].isna().sum()
````

````python
# 6) Seu código aqui
sinasc['QTDFILVIVO'] = sinasc['QTDFILVIVO'].fillna(0)
sinasc['QTDFILVIVO'].isna().sum()
````

````python
# 7) seu código aqui
sinasc[colunas].isna().sum()





# As variaveis faltantes possuem indicacao na tabela do que fazer caso haja um dado faltante (preencher com '9 : Ignorado');
# os dados em questão são valores que não podem ser aproximados, pois seria necessário outros tipos de referência ou dados que não 
# podem ser preenchidos pelo senso-comum (como o tipo de gestação ou o número de semanas de gestação)
````

````python
sinasc[['ESCMAE','GESTACAO','GRAVIDEZ']] = sinasc[['ESCMAE','GESTACAO','GRAVIDEZ']].fillna(9)
sinasc[colunas].isna().sum()
````

````python
# 8) seu código aqui
sinasc.loc[(sinasc['APGAR5'] >= 8) & (sinasc['APGAR5'] <= 10), 'cat apgar5'] = 'normal'
sinasc.loc[(sinasc['APGAR5'] >= 6) & (sinasc['APGAR5'] <= 7), 'cat apgar5'] = 'asfixia leve'
sinasc.loc[(sinasc['APGAR5'] >= 4) & (sinasc['APGAR5'] <= 5), 'cat apgar5'] = 'asfixia moderada'
sinasc.loc[(sinasc['APGAR5'] >= 0) & (sinasc['APGAR5'] <= 3), 'cat apgar5'] = 'asfixia severa'

#deixei cat apgar5 com espaço para exercitar o próximo exercício, para ver se consegui substituir o espaço por _
sinasc.head(20)
````
